```python
import pandas as pd 
```

```python
df = pd.read_csv('test_case1.csv') # Импортируем данные из таблицы
```

```python
df
```

```python
df_ranks = df.drop(columns=['geo', 'device', 'metric']) # Для удобства сначала отдельно возьмём параметры с численными значениями.
df_ranks.replace('-', 0, inplace=True) # Так как первые три дня даны в формате строк, преобразуем отсутсвующие значения в нули, чтобы не мешали превращению всей таблицы в численный формат.
df_ranks.astype('int64') # Трансформируем таблицу в численный формат.
df_ranks = df_ranks.map(lambda x: 1 if x in [1, 2, 3] else 0) # Накладываем "маску" на таблицу, чтобы считались места только в топ-3, причём считались за 1, чтобы по рядам легко посчитать сумму.
df = pd.concat([df[['geo', 'device', 'metric']], df_ranks], axis=1) # Соединим преобразованную таблицу с данными регионов и устройств.
df
```

```python
df = df[df.drop(columns=['geo', 'device', 'metric']).sum(axis=1) >= 3].drop(columns=['metric']) # Оставляем только ряды, отвечающие условию: минимум 3 попадания в топ-3.
df
```

```python
df.groupby(['device']).count().iloc[:,0] # Группируем по виду устройств и смотрим количество совпадений.
```
